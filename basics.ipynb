## A sample session

Ref: https://cran.r-project.org/doc/manuals/r-release/R-intro.html#A-sample-session

```r
# generate two pseudo-random normal vectors of x- and y-coordinates.
x <- rnorm(50)
y <- rnorm(x)
plot(x,y)    # plot points in the plane
ls()         # list R objects in R workspace
rm(x,y)      # delete objects
```

```r
x <- 1:10             # make x = (1, 2, ..., 10).
w <- 1 + sqrt(x)/2    # A ‘weight’ vector of standard deviations
dummy <- data.frame(x=x, y=x+rnorm(x)*w)  # make a data frame of two columns
dummy                 # show it
```

```r
# fit a simple linear regression and look at the analysis
# y to the left of the tilde: we are modelling y dependent on x
fm <- lm(y ~ x, data=dummy)
summary(fm)
```
