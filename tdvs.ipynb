## MPG

From `?mpg`:  This dataset contains a subset of the fuel economy data that the EPA makes available on <https://fueleconomy.gov/>. It contains only models which had a new release every year between 1999 and 2008 - this was used as a proxy for the popularity of the car.

```r
# The data we are going to work with (to easy flipping)
theData = mpg  # fuel economy
#?mpg
```

```r
#theData  # like View(theData), show(theData), ...
#head(theData)
glimpse(theData)
```

```r
View(length(theData))
theData %>% length()
```

```r
View(length(theData$cty))
```

```r
View(theData$cty)
```

```r
View(mean(theData$cty, na.rm=T))
```

```r
#filter(theData, cty >= 25)
#theData %>% filter(cty >=25)
theData2 <- theData %>% filter(cty >= 25)
View(theData2)
mean(theData2$hwy)
```

```r
theData3 <- theData2 %>% mutate(cty_metric = 0.425144 * cty)
glimpse(theData3)
View(theData3)
```

```r
theData3 %>%
  group_by(class) %>%
  summarise(
    mean(cty_metric),
    median(cty_metric)
  )
```

```r
theData3 %>%
  ggplot(aes(x = cty_metric, y = hwy)) +
  geom_point(size=5) +
  geom_line() +
  geom_smooth(method = "lm") +
  labs(title = "CTY Metric", x = "CTY Metric", y = "Total")
```
