## Vector → Data frame

```r
my_vector <- c(1:12)

# Fill by ROW (1, 2, 3, 4 will be the first row)
matrix_by_row <- matrix(my_vector, ncol = 4, byrow = TRUE)
df_row <- data.frame(matrix_by_row)
colnames(df_row) <- c("Col1", "Col2", "Col3", "Col4")

# Fill by COLUMN (1, 2, 3 will be the first column)
matrix_by_col <- matrix(my_vector, ncol = 4, byrow = FALSE)
df_col <- data.frame(matrix_by_col)
colnames(df_col) <- c("Col1", "Col2", "Col3", "Col4")

my_vector
df_row
df_col
```

## A sample session

Ref: https://cran.r-project.org/doc/manuals/r-release/R-intro.html#A-sample-session

```r
# generate two pseudo-random normal vectors of x- and y-coordinates.
x <- rnorm(50)
y <- rnorm(x)
plot(x,y)    # plot points in the plane
ls()         # list R objects in R workspace
rm(x,y)      # delete objects
```

```r
x <- 1:10             # make x = (1, 2, ..., 10).
w <- 1 + sqrt(x)/2    # A ‘weight’ vector of standard deviations
dummy <- data.frame(x=x, y=x+rnorm(x)*w)  # make a data frame of two columns
dummy                 # show it
```

```r
# fit a simple linear regression and look at the analysis
# y to the left of the tilde: we are modelling y dependent on x
fm <- lm(y ~ x, data=dummy)
summary(fm)
```

```r
library(tidyverse)  # glipmse(), %>%, mpg, ...
```

```r
# The data we are going to work with (to easy flipping)
theData = mpg
#?mpg
```

```r
theData  # like View(theData), show(theData), ...
```

```r
View(length(theData))
View(theData$cty)
View(mean(theData$cty, na.rm=T))
```

```r
glimpse(theData)
length(theData)
theData %>% length()
```

```r
#filter(theData, cty >= 25)
#theData %>% filter(cty >=25)
theData2 <- theData %>% filter(cty >= 25)
View(theData2)
mean(theData2$hwy)
```

```r
theData3 <- theData2 %>% mutate(cty_metric = 0.425144 * cty)
glimpse(theData3)
View(theData3)
```

```r
theData3 %>%
  group_by(class) %>%
  summarise(
    mean(cty_metric),
    median(cty_metric)
  )
```

```r
theData3 %>%
  ggplot(aes(x = cty_metric, y = hwy)) +
  geom_point(size=5) +
  geom_line() +
  geom_smooth(method = "lm") +
  labs(title = "CTY Metric", x = "CTY Metric", y = "Total")
```
